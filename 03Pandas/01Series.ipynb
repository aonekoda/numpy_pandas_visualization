# Pandas Series

파이썬 시리즈(Python Series)는 판다스의 중요 자료구조 중 하나로 1차원 자료구조이다. 

````
import pandas as pd
import numpy as np
pd.__version__
````

````
s = pd.Series([1, 3, 5, np.nan, 1, 2], index=['A','B','C','D','E','F'])
s
````

### 인덱싱과 슬라이싱

numpy 배열과 같은 방법으로 인덱싱과 슬라이싱 할 수 있다.

````
s[0]
````

````
s[2:5]
````

````
s['A']
````

````
s['A':'C']
````

### 시리즈의 메소드

````
s.value_counts() # 고유값의 빈도수
````

````
s.index # 인덱스 이름
````

````
s.shape
````

````
s.count()  # 결측치를 제외한 갯수
````

````
s.sort_values()
````

````
s.unique()
````

````
s.nunique()
````

````
s.isin([1,2])
````

````
s.min()  
````

````
s.describe()
````

````
s.quantile(q=0.75)
````

````
s.isnull()
````

````
s.fillna(-99)
````

````
s.dropna()
````

````
s1 = pd.Series([10,20,30,40])
s2 = pd.Series([1,2,300,400])
````

````
s1 + s2
````

````
s1 > s2
````

````
s1[s1 > s2]
````
